# 1.  Generate a list of 100 integers containing values between 90 to 130 and store it in the variable `int_list`.

Create a function to find the median of a list of numbers.

  

  (ii) Develop a program to compute the mode of a list of integers.

  

  (iii) Implement a function to calculate the weighted mean of a list of values and their corresponding weights.

  

  (iv) Write a Python function to find the geometric mean of a list of positive numbers.

  

  (v) Create a program to calculate the harmonic mean of a list of values.

  

  (vi) Build a function to determine the midrange of a list of numbers (average of the minimum and maximum).

  

  (vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of
outliers.

````python
# Answer
import random
import statistics
from scipy.stats import trim_mean
import numpy as np

# Step 1: Generate a list of 100 random integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

# Step 2: Functions to calculate various means

# (i) Function to calculate the mean
def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

# Function to find the median
def calculate_median(numbers):
    return statistics.median(numbers)

# (ii) Function to calculate the mode
def calculate_mode(numbers):
    return statistics.mode(numbers)

# (iii) Function to calculate the weighted mean
def calculate_weighted_mean(values, weights):
    return sum(value * weight for value, weight in zip(values, weights)) / sum(weights)

# (iv) Function to calculate the geometric mean
def calculate_geometric_mean(numbers):
    return np.prod(np.array(numbers)) ** (1 / len(numbers))

# (v) Function to calculate the harmonic mean
def calculate_harmonic_mean(numbers):
    return len(numbers) / sum(1 / x for x in numbers)

# (vi) Function to calculate the midrange
def calculate_midrange(numbers):
    return (max(numbers) + min(numbers)) / 2

# (vii) Function to calculate the trimmed mean
def calculate_trimmed_mean(numbers, percentage):
    return trim_mean(numbers, proportiontocut=percentage / 100)

# Example usage of the functions
mean = calculate_mean(int_list)
median = calculate_median(int_list)
mode = calculate_mode(int_list)
weighted_mean = calculate_weighted_mean(int_list, [1]*len(int_list))  # Example weights
geometric_mean = calculate_geometric_mean(int_list)
harmonic_mean = calculate_harmonic_mean(int_list)
midrange = calculate_midrange(int_list)
trimmed_mean = calculate_trimmed_mean(int_list, 10)  # Exclude 10% of outliers

# Display the results
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Weighted Mean: {weighted_mean}")
print(f"Geometric Mean: {geometric_mean}")
print(f"Harmonic Mean: {harmonic_mean}")
print(f"Midrange: {midrange}")
print(f"Trimmed Mean (10%): {trimmed_mean}")
````

````python
# 3  Write a Python class representing a discrete random variable with methods to calculate its expected value and variance.

class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        """
        Initialize the discrete random variable.
        :param values: A list of possible values.
        :param probabilities: A list of corresponding probabilities.
        """
        if len(values) != len(probabilities):
            raise ValueError("The length of values and probabilities must be the same.")
        if not np.isclose(sum(probabilities), 1.0):
            raise ValueError("The sum of the probabilities must be 1.")
        
        self.values = values
        self.probabilities = probabilities

    def expected_value(self):
        """
        Calculate the expected value (mean) of the discrete random variable.
        :return: Expected value (float)
        """
        return sum(value * prob for value, prob in zip(self.values, self.probabilities))

    def variance(self):
        """
        Calculate the variance of the discrete random variable.
        :return: Variance (float)
        """
        mean = self.expected_value()
        return sum(prob * (value - mean) ** 2 for value, prob in zip(self.values, self.probabilities))

# Example usage:

# Define the values and their corresponding probabilities
values = [1, 2, 3, 4, 5]
probabilities = [0.1, 0.2, 0.3, 0.25, 0.15]

# Create an instance of the DiscreteRandomVariable class
drv = DiscreteRandomVariable(values, probabilities)

# Calculate the expected value and variance
expected_value = drv.expected_value()
variance = drv.variance()

print(f"Expected Value: {expected_value}")
print(f"Variance: {variance}")
````

````python
# 4. Implement a program to simulate the rolling o6 a 6air six-sided die and calculate the expected value and variance o6 the outcomes.
import random

class DiceSimulation:
    def __init__(self, num_rolls):
        """
        Initialize the simulation with the number of rolls.
        :param num_rolls: Number of times to roll the dice.
        """
        self.num_rolls = num_rolls
        self.outcomes = []

    def roll_dice(self):
        """
        Simulate rolling a pair of six-sided dice.
        :return: Sum of the dice (int)
        """
        die1 = random.randint(1, 6)
        die2 = random.randint(1, 6)
        return die1 + die2

    def simulate(self):
        """
        Simulate rolling the dice multiple times and store the outcomes.
        """
        self.outcomes = [self.roll_dice() for _ in range(self.num_rolls)]

    def expected_value(self):
        """
        Calculate the expected value (mean) of the outcomes.
        :return: Expected value (float)
        """
        return sum(self.outcomes) / len(self.outcomes)

    def variance(self):
        """
        Calculate the variance of the outcomes.
        :return: Variance (float)
        """
        mean = self.expected_value()
        return sum((outcome - mean) ** 2 for outcome in self.outcomes) / len(self.outcomes)

# Example usage:

# Number of times to roll the dice
num_rolls = 10000

# Create an instance of DiceSimulation
dice_simulation = DiceSimulation(num_rolls)

# Simulate the dice rolls
dice_simulation.simulate()

# Calculate the expected value and variance
expected_value = dice_simulation.expected_value()
variance = dice_simulation.variance()

print(f"Expected Value: {expected_value}")
print(f"Variance: {variance}")
````

````python
# 5. Create a Python 6unction to generate random samples 6rom a given probability distribution (e.g., binomial, Poisson) and calculate their mean and variance.
import numpy as np

def generate_samples_and_calculate_stats(distribution, params, size=1000):
    """
    Generate random samples from a specified probability distribution and calculate their mean and variance.
    
    :param distribution: A string representing the distribution ('binomial', 'poisson', etc.).
    :param params: A dictionary of parameters required by the distribution.
    :param size: Number of samples to generate (default is 1000).
    :return: A tuple containing the mean and variance of the generated samples.
    """
    if distribution == 'binomial':
        # params should contain 'n' (number of trials) and 'p' (probability of success)
        samples = np.random.binomial(n=params['n'], p=params['p'], size=size)
    elif distribution == 'poisson':
        # params should contain 'lam' (lambda, the rate or expected number of occurrences)
        samples = np.random.poisson(lam=params['lam'], size=size)
    elif distribution == 'normal':
        # params should contain 'mu' (mean) and 'sigma' (standard deviation)
        samples = np.random.normal(loc=params['mu'], scale=params['sigma'], size=size)
    else:
        raise ValueError("Unsupported distribution type")

    mean = np.mean(samples)
    variance = np.var(samples)

    return mean, variance

# Example usage:

# Generate samples from a Binomial distribution with n=10 trials and probability p=0.5
binomial_params = {'n': 10, 'p': 0.5}
binomial_mean, binomial_variance = generate_samples_and_calculate_stats('binomial', binomial_params)
print(f"Binomial Distribution - Mean: {binomial_mean}, Variance: {binomial_variance}")

# Generate samples from a Poisson distribution with lambda=3
poisson_params = {'lam': 3}
poisson_mean, poisson_variance = generate_samples_and_calculate_stats('poisson', poisson_params)
print(f"Poisson Distribution - Mean: {poisson_mean}, Variance: {poisson_variance}")

# Generate samples from a Normal distribution with mean=0 and standard deviation=1
normal_params = {'mu': 0, 'sigma': 1}
normal_mean, normal_variance = generate_samples_and_calculate_stats('normal', normal_params)
print(f"Normal Distribution - Mean: {normal_mean}, Variance: {normal_variance}")
````

````python
#6. Write a Python script to generate random numbers 6rom a Gaussian (normal) distribution and compute the mean, variance, and standard deviation o6 the samples.
import numpy as np

def generate_normal_samples(mean, std_dev, size=1000):
    """
    Generate random samples from a normal (Gaussian) distribution and compute their statistics.
    
    :param mean: Mean of the normal distribution.
    :param std_dev: Standard deviation of the normal distribution.
    :param size: Number of samples to generate (default is 1000).
    :return: A tuple containing the mean, variance, and standard deviation of the generated samples.
    """
    # Generate random samples from the normal distribution
    samples = np.random.normal(loc=mean, scale=std_dev, size=size)
    
    # Compute the mean, variance, and standard deviation
    sample_mean = np.mean(samples)
    sample_variance = np.var(samples)
    sample_std_dev = np.std(samples)
    
    return sample_mean, sample_variance, sample_std_dev

# Example usage:

# Parameters for the normal distribution
mean = 0      # Mean of the distribution
std_dev = 1   # Standard deviation of the distribution
size = 1000   # Number of samples to generate

# Generate samples and compute statistics
mean_value, variance_value, std_dev_value = generate_normal_samples(mean, std_dev, size)

# Print the results
print(f"Mean: {mean_value}")
print(f"Variance: {variance_value}")
print(f"Standard Deviation: {std_dev_value}")
````

````python
# 8. Write a Python function to calculate the probability density function (PDF) of a continuous random variable for a given normal distribution.
import scipy.stats as stats

def normal_pdf(x, mean=0, std_dev=1):
    """
    Calculate the Probability Density Function (PDF) of a normal distribution.
    
    :param x: The value(s) at which to evaluate the PDF.
    :param mean: The mean of the normal distribution (default is 0).
    :param std_dev: The standard deviation of the normal distribution (default is 1).
    :return: The PDF value(s) at the specified x.
    """
    return stats.norm.pdf(x, loc=mean, scale=std_dev)

# Example usage:

# Define the parameters for the normal distribution
mean = 0      # Mean of the distribution
std_dev = 1   # Standard deviation of the distribution

# Evaluate the PDF at specific values
x_values = [0, 1, 2]
pdf_values = [normal_pdf(x, mean, std_dev) for x in x_values]

print(f"PDF values at x={x_values}: {pdf_values}")
````

# 11. A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors who make a purchase). They collect data from the old and new layouts to compare.

To generate the data use the following command:

```python
import numpy as np
# 50 purchases out of 1000 visitors
old_layout = np.array([1] * 50 + [0] * 950)
# 70 purchases out of 1000 visitors  
new_layout = np.array([1] * 70 + [0] * 930)
 ```
Apply z-test to find which layout is successful.

````python
# Answer :- 
import numpy as np
import scipy.stats as stats

# Data
old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)

# Number of successes and trials
successes_old = np.sum(old_layout)
successes_new = np.sum(new_layout)
n_old = len(old_layout)
n_new = len(new_layout)

# Proportions
p1 = successes_old / n_old
p2 = successes_new / n_new

# Pooled proportion
p_pooled = (successes_old + successes_new) / (n_old + n_new)

# Standard error of the difference in proportions
se = np.sqrt(p_pooled * (1 - p_pooled) * (1 / n_old + 1 / n_new))

# Z-statistic
z = (p1 - p2) / se

# P-value for the two-tailed test
p_value = 2 * (1 - stats.norm.cdf(np.abs(z)))

# Output results
print(f"Proportion of successes in old layout: {p1:.4f}")
print(f"Proportion of successes in new layout: {p2:.4f}")
print(f"Z-statistic: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the p-value
alpha = 0.05
if p_value < alpha:
    print("The difference in conversion rates is statistically significant. The new layout is successful.")
else:
    print("The difference in conversion rates is not statistically significant. No conclusive evidence that the new layout is better.")
````

# 12. A tutoring service claims that its program improves students' exam scores. A sample oN students who participated in the program was taken, and their scores beNore and aNter the program were recorded.
Use the below code to generate samples of respective arrays of marks:

```python

before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])

after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

```

Use z-test to find if the claims made by tutor are true or false.

````python
# Answer :- 
import numpy as np
import scipy.stats as stats

# Data
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

# Calculate the differences
differences = after_program - before_program

# Mean and standard deviation of differences
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=0)  # Population standard deviation

# Number of observations
n = len(differences)

# Standard error of the mean difference
se = std_diff / np.sqrt(n)

# Z-statistic
z = mean_diff / se

# P-value for the one-tailed test (since we're testing if the program improves scores)
p_value = 1 - stats.norm.cdf(z)

# Output results
print(f"Mean difference in scores: {mean_diff:.4f}")
print(f"Standard deviation of differences: {std_diff:.4f}")
print(f"Standard error of the mean difference: {se:.4f}")
print(f"Z-statistic: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the p-value
alpha = 0.05
if p_value < alpha:
    print("The improvement in scores is statistically significant. The tutoring program is effective.")
else:
    print("The improvement in scores is not statistically significant. No conclusive evidence that the tutoring program is effective.")
````

Write a Python function to perfor: an A/B test analysis, including calculating the t-statistic, degrees of
freedo:, and p-value.


Use the followi3g data:

```python

layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]

layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

````python
# Answer 
import numpy as np
import scipy.stats as stats

def ab_test_analysis(layout_a_clicks, layout_b_clicks):
    # Convert lists to numpy arrays
    a = np.array(layout_a_clicks)
    b = np.array(layout_b_clicks)

    # Calculate sample statistics for both layouts
    mean_a = np.mean(a)
    std_a = np.std(a, ddof=1)  # Sample standard deviation
    n_a = len(a)

    mean_b = np.mean(b)
    std_b = np.std(b, ddof=1)  # Sample standard deviation
    n_b = len(b)

    # Standard error of the difference between means
    se = np.sqrt((std_a**2 / n_a) + (std_b**2 / n_b))

    # T-statistic
    t_statistic = (mean_a - mean_b) / se

    # Degrees of freedom (Welch-Satterthwaite equation)
    df = ( (std_a**2 / n_a + std_b**2 / n_b)**2 /
           ( ( (std_a**2 / n_a)**2 / (n_a - 1) ) + ( (std_b**2 / n_b)**2 / (n_b - 1) ) )
         )

    # P-value for the two-tailed test
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df=df))

    return {
        'mean_a': mean_a,
        'std_a': std_a,
        'mean_b': mean_b,
        'std_b': std_b,
        't_statistic': t_statistic,
        'degrees_of_freedom': df,
        'p_value': p_value
    }

# Data
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Perform A/B test analysis
results = ab_test_analysis(layout_a_clicks, layout_b_clicks)

# Output results
print(f"Layout A - Mean: {results['mean_a']:.2f}, Std Dev: {results['std_a']:.2f}")
print(f"Layout B - Mean: {results['mean_b']:.2f}, Std Dev: {results['std_b']:.2f}")
print(f"T-statistic: {results['t_statistic']:.4f}")
print(f"Degrees of Freedom: {results['degrees_of_freedom']:.2f}")
print(f"P-value: {results['p_value']:.4f}")

# Interpret the p-value
alpha = 0.05
if results['p_value'] < alpha:
    print("There is a statistically significant difference between the click-through rates of the two layouts.")
else:
    print("There is no statistically significant difference between the click-through rates of the two layouts.")
````

# 21.A political analyst wants to determine if there is a significant association between age groups and voter
preferences FCandidate A or Candidate B). They collect data from a sample of 500 voters and classify
them into different age groups and candidate preferences. Perform a Chi-Square test to determine if
there is a significant association between age groups and voter preferences.


Use the below code to generate data:

```python

np.random.seed(0)

age_groups = np.random.choice([ 18 30 , 31 50 , 51+', 51+'], size=30)

voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

````python
# Answer
import numpy as np
import pandas as pd
import scipy.stats as stats

def chi_square_test(age_groups, voter_preferences):
    # Create a DataFrame from the data
    data = pd.DataFrame({'Age Group': age_groups, 'Voter Preference': voter_preferences})

    # Generate the contingency table
    contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

    # Perform the Chi-Square test
    chi2_statistic, p_value, dof, expected = stats.chi2_contingency(contingency_table)

    return {
        'contingency_table': contingency_table,
        'chi2_statistic': chi2_statistic,
        'degrees_of_freedom': dof,
        'p_value': p_value,
        'expected_frequencies': expected
    }

# Generate synthetic data
np.random.seed(0)
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=500)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)

# Perform the Chi-Square test
results = chi_square_test(age_groups, voter_preferences)

# Output results
print("Contingency Table:")
print(results['contingency_table'])
print(f"\nChi-Square Statistic: {results['chi2_statistic']:.4f}")
print(f"Degrees of Freedom: {results['degrees_of_freedom']:.2f}")
print(f"P-value: {results['p_value']:.4f}")

# Interpret the p-value
alpha = 0.05
if results['p_value'] < alpha:
    print("There is a statistically significant association between age groups and voter preferences.")
else:
    print("There is no statistically significant association between age groups and voter preferences.")
````
